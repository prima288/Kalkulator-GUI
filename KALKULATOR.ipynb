```python
from tkinter import *
import math

window = Tk()
window.title("Kalkulator GUI Dengan Python")
window.geometry('500x500')
window.configure(bg='#FFFFFF')

header_label = Label(window, text="Kalkulator", font=("Arial Bold", 20), bg='#FFFFFF')
header_label.grid(columnspan=2, row=0, padx=20, pady=20)

nilai1_label = Label(window, text="Masukkan Nilai Pertama:", font=("Arial Bold", 12), bg='#FFFFFF')
nilai1_label.grid(column=0, row=1, padx=10, pady=10)
nilai1_entry = Entry(window, width=20, font=("Arial", 12), bg="#f0f0f0")
nilai1_entry.grid(column=1, row=1, padx=10, pady=10)

nilai2_label = Label(window, text="Masukkan Nilai Kedua:", font=("Arial Bold", 12), bg='#FFFFFF')
nilai2_label.grid(column=0, row=2, padx=10, pady=10)
nilai2_entry = Entry(window, width=20, font=("Arial", 12), bg="#f0f0f0")
nilai2_entry.grid(column=1, row=2, padx=10, pady=10)

hasil_label = Label(window, text="Hasil:", font=("Arial Bold", 12), bg='#FFFFFF')
hasil_label.grid(column=0, row=3, padx=10, pady=10)
hasil = Label(window, text="0", font=("Arial", 12), anchor="w", width=10, bg="#f0f0f0")
hasil.grid(column=1, row=3, padx=10, pady=10)

def tambah():
    hasil.configure(text=(float(nilai1_entry.get())+float(nilai2_entry.get())))

def kurang():
    hasil.configure(text=(float(nilai1_entry.get())-float(nilai2_entry.get())))

def kali():
    hasil.configure(text=(float(nilai1_entry.get())*float(nilai2_entry.get())))

def bagi():
    hasil.configure(text=(float(nilai1_entry.get())/float(nilai2_entry.get())))

def pangkat():
    if float(nilai2_entry.get()) >= 0:
        hasil.configure(text=(float(nilai1_entry.get())**(float(nilai2_entry.get()))))
    else:
        hasil.configure(text="Error: Invalid input: Nilai pangkat tidak boleh negatif")

def modulus():
    hasil.configure(text=(float(nilai1_entry.get())%float(nilai2_entry.get())))

def akar():
    if float(nilai2_entry.get()) >= 0:
        hasil.configure(text=(float(nilai1_entry.get())**(1/float(nilai2_entry.get()))))
    else:
        hasil.configure(text="Error: Invalid input: Nilai pangkat tidak boleh negatif")


def pangkat():
    if float(nilai2_entry.get()) >= 0 or float(nilai1_entry.get()) >= 0:
        hasil.configure(text=(float(nilai1_entry.get())**(float(nilai2_entry.get()))))
    else:
        hasil.configure(text="Error: Invalid input: Nilai dasar dan pangkat tidak boleh negatif")


tambah_button = Button(window, text="Tambah", font=("Helvetica", 12), command=tambah, bg="#89CFF0", fg="white")
tambah_button.grid(column=0, row=4, padx=10, pady=10)

kurang_button = Button(window, text="Kurang", font=("Helvetica", 12), command=kurang, bg="#89CFF0", fg="white")
kurang_button.grid(column=1, row=4, padx=10, pady=10)

kali_button = Button(window, text="Kali", font=("Helvetica", 12), command=kali, bg="#89CFF0", fg="white")
kali_button.grid(column=0, row=5, padx =10, pady=10)

bagi_button = Button(window, text="Bagi", font=("Helvetica", 12), command=bagi, bg="#89CFF0", fg="white")
bagi_button.grid(column=1, row=5, padx=10, pady=10)

modulus_button = Button(window, text="Modulus", font=("Helvetica", 12), command=modulus, bg="#89CFF0", fg="white")
modulus_button.grid(column=0, row=6, padx=10, pady=10)

akar_button = Button(window, text="Akar", font=("Helvetica", 12), command=akar, bg="#89CFF0", fg="white")
akar_button.grid(column=1, row=6, padx=10, pady=10)

pangkat_button = Button(window, text="Pangkat", font=("Helvetica", 12), command=pangkat, bg="#89CFF0", fg="white")
pangkat_button.grid(column=0, row=7, padx=10, pady=10)


window.mainloop()
```
